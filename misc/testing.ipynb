```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read File
data = pd.read_csv("data.csv")

# Declare Data
values = data['amount']
labels = data['category']
mycolors = ['#49642a', '#606c38','#aabf64', '#bc6c25', '#dda15e', '#cdcab4'] 

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format

# Plot
fig, ax = plt.subplots(figsize=(5, 4))
# plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

ax.pie(values, 
       labels=labels, 
       colors = mycolors, 
       autopct = autopct_format(values), 
       pctdistance=0.8,
       textprops={'fontsize': 10, 'color': 'black', 'family': 'Montserrat'})

centre_circle = plt.Circle((0, 0), 0.60, fc="white")
ax.add_artist(centre_circle)
# Add Title (Centered)
plt.title("Testing title report blablah\nreport 3: balbla\n", fontsize=14, fontweight="bold", ha="center")

# Ensure the pie remains circular
ax.axis("equal")

# Adjust layout for better centering
fig.tight_layout()

# Save the figure (Ensuring all elements fit properly)
plt.savefig("figure.png", bbox_inches='tight')
```
